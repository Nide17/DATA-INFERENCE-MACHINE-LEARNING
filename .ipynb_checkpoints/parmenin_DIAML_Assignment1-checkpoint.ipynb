### Question 3

```python
# Formula: A = P(1+(r/n))**nt
P = 100
r = 5/100
n = 1 #Frequency is 1 since the annualized interest rate is compounded annually

# Finding the amount after years: Amount= P * pow((1+r/n), n * t) 
for t in [1, 2, 3, 4, 5]:
    amount= P * pow((1 + (r / n)), (n * t))
    print(round(amount))
```

### Question 4

```python
#  Loan payment formula: M = Pr / (1 −(1+r) ** −n)

# M: is the monthly payment
# P: is the original principal amount
# r: is the interest rate per month (annual interest divided by 12)
# n: is the number of months to repay the loan or total number of payments

P = 20000
r = 0.01 # the 1% interest rate

for yearNbr in [1, 2, 3]:
    M = round((P * r) / (1 - (pow((1 + r), (-12 * yearNbr))))) # Applying the "M = Pr / (1 −(1+r) ** −n)" formula
    print(M)
```

### Question 5

```python
initInvest = 100000
custNbr = 100

pymts = 0
acPymtsList = []

daysNbr = 0
daysList = []

while pymts <= initInvest:                       # iterate until the payments is greater than or equal to amount invested
    custNbr = custNbr + (custNbr * (1 / 100))    # updating customers
    daysNbr = daysNbr + 1                        # updating days
    daysList.append(daysNbr)                     # Keeping track of the number of days
    pymts = pymts + custNbr * 10                 # updating the payments
    acPymtsList.append(pymts)                    # Keeping track of accumulated profit for each day
    
print(daysNbr)
```

```python
# Plotting cumulated profits per day, show initial investment and mark breakeven day. === matplotlib
plt.figure(figsize=(10, 6))
plt.plot(daysList, acPymtsList)

plt.title("CUMULATED PROFITS PER DAY")
plt.xlabel("Days")
plt.ylabel("Payments") 

profitDay = np.interp(initInvest, acPymtsList , daysList)

plt.scatter(profitDay, initInvest, color = 'red', s=100, marker = '*')  # Showing the break even point

plt.show()
```

### Question 6

```python
ebolaExcelData = pd.read_excel('ebola_download.xls')                       # Reading the data from excel file
ebolaExtractedData = ebolaExcelData[['Date','Cases','Death']]              # Extracting Dates, Cases, and Deaths
ebolaExtractedData.info()                                                  # Getting info about the data 
```

```python
# Get missing dates by generating dates from the first date to the last date in the dataset
ebolaSortedData = ebolaExtractedData.sort_values(by='Date')                   # Sorting the data by date first
firstDate = ebolaSortedData.Date[0]                                           # The first date in the sorted data
lastDate = ebolaSortedData.Date[len(ebolaExtractedData['Date']) - 1]          # The last date in the sorted data

allDates = pd.date_range(start = firstDate, end = lastDate)                   # Filling in the skipped dates into the range
len(allDates)                                                                 # View the length or amount of total dates
```

```python
ebolaSortedData = ebolaSortedData.set_index('Date')                           # Set the 'Date' column to be the index of dataset

# Integrate the missing dates into dataset. Conform dataset to new index. Placing NaN where no value exist 
ebolaReindexedData = ebolaSortedData.reindex(allDates)  
ebolaReindexedData.isnull().sum()                                             # Counting empty fields
```

```python
ebolaInterpolatedData = ebolaReindexedData.interpolate()                        # Interpolating linearly to get NaN (missing) values for cases and deaths
ebolaInterpolatedData = ebolaInterpolatedData.rename_axis('Date').reset_index() # Adding the title to the index (date) column
ebolaInterpolatedData.isnull().sum()                                            # Counting  empty fields
```

#### Extracting indices (first index) where case values are greater than 100, 500, 1000, 2000, 5000

##### Ebola cases values

```python
casesList = [100, 500, 1000, 2000, 5000]
i = 0
excCases = []

while i < len(casesList):
    for c in ebolaInterpolatedData['Cases']:
        if c >= casesList[i]: 
            break
    excCases.append(c)
    i = i + 1
    
excCases
```

##### Ebola cases dates values

```python
c = 0
excCasesDates = []

while c < len(excCases):                                              
    caseDate = ebolaInterpolatedData[ebolaInterpolatedData['Cases'] == excCases[c]]
    caseDate = caseDate['Date']
    
    print(caseDate.iloc[0])                                                        # Printing 5 dates of ebola cases
    excCasesDates.append(caseDate)                                                 # Updating excCasesDates list
    c = c + 1
```

##### Ebola deaths values

```python
deathsList = [100, 500, 1000, 2000, 5000]
i = 0
excDeaths = []

while i < len(deathsList):
    for c in ebolaInterpolatedData['Death']:
        if c >= deathsList[i]: 
            break
    excDeaths.append(c)
    i = i + 1
    
excDeaths
```

##### Ebola cases dates values

```python
d = 0
excDeathsDates = []

while d < len(excDeaths):
    deathDate = ebolaInterpolatedData[ebolaInterpolatedData['Death'] == excDeaths[d]]
    deathDate = deathDate['Date']
    
    print(deathDate.iloc[0])                                                        # Printing 5 dates of ebola deaths
    excDeathsDates.append(deathDate)                                                # Updating excDeathsDates list
    d = d + 1
```

#### Plotting cases vs dates

```python
ebolaCasesData = ebolaInterpolatedData["Cases"]
ebolaDeathsData = ebolaInterpolatedData["Death"]

plt.figure(figsize = (10, 8))

plt.plot(ebolaInterpolatedData.Date, ebolaCasesData, label = 'Cases')
plt.scatter(excCasesDates, excCases)

plt.plot(ebolaInterpolatedData.Date, ebolaDeathsData, label = 'Deaths')
plt.scatter(excDeathsDates, excDeaths, marker = 'o')

plt.title("GRAPH OF EBOLA CASES AND DEATHS AGAINST DATES")
plt.xlabel("Dates")
plt.ylabel("Values of Ebola Cases and Deaths")
plt.legend()
plt.show()
```

### Question 7

```python
interpolatedCases = ebolaInterpolatedData['Cases']              # Getting interpolated cases from question 6
interpolatedDeaths = ebolaInterpolatedData['Death']             # Getting interpolated deaths from question 6
```

```python
# Computing growth rates: Formula: GR = (Absolute change / Original value) * 100 => Absolute change = New value - Original value

casesGR = 0
c = 1 

while c <= (len(interpolatedCases) - 1):
    DailyGR = ((interpolatedCases[c] - interpolatedCases[c-1]) / interpolatedCases[c-1]) * 100     # The daily growth rate
    casesGR = casesGR + DailyGR
    c = c + 1

casesAvgGR = casesGR / (len(interpolatedCases) - 1)             # The average growth rate for cases

print(casesAvgGR)
```

```python
# Computing growth rates: Formula: GR = (Absolute change / Original value) * 100 => Absolute change = New value - Original value
deathsGR = 0
d = 1 

while d <= (len(interpolatedDeaths) - 1):
    DailyGR = ((interpolatedDeaths[d] - interpolatedDeaths[d-1]) / interpolatedDeaths[d-1]) * 100
    deathsGR = deathsGR + DailyGR
    d = d + 1

deathsAvgGR = deathsGR / (len(interpolatedDeaths) - 1)           # The average growth rate for deaths

print(deathsAvgGR)
```

### Question 9

```python
# Reading data
tlt = pd.read_csv('TLT.csv')
spy = pd.read_csv('SPY.csv')

# Extracting dates and adjusted closing values
extractedTLT = tlt[['Date', 'Adj Close']] 
extractedSPY = spy[['Date', 'Adj Close']] 
```

```python
# Normalizing so that the first value for each stock is 100 
extractedAdjCloseTLT = extractedTLT['Adj Close']
extractedAdjCloseSPY = extractedSPY['Adj Close']

normalizedTLT = []
normalizedSPY = []

for tlt in extractedAdjCloseTLT:
    normalized = (tlt / extractedAdjCloseTLT[0]) * 100
    normalizedTLT.append(normalized)
    
for spy in extractedAdjCloseSPY:
    normalized = (spy / extractedAdjCloseSPY[0]) * 100
    normalizedSPY.append(normalized)

extractedTLT['Date']
```

```python
# Plotting

extractedTLT['Date'] = pd.to_datetime(extractedTLT['Date'])              # converting the timestamps time to datetime
extractedSPY['Date'] = pd.to_datetime(extractedSPY['Date'])

plt.xlim(pd.Timestamp('2014-01-01'), pd.Timestamp('2015-08-31'))         # set time limits of the Date axis.

plt.plot(extractedTLT['Date'], normalizedTLT, label = 'TLT')
plt.plot(extractedSPY['Date'], normalizedSPY, label = 'SPY')

plt.ylabel("TLT and SPY")
plt.xlabel("Date")
plt.legend()
plt.show()
```

### Question 10

```python
dailyTLTReturnList = []
dailySPYReturnList = []

for t in range(1, len(normalizedTLT)):
    dailyReturnT = ((normalizedTLT[t] / normalizedTLT[t-1]) -1) * 100
    dailyTLTReturnList.append(dailyReturnT)
    
for s in range(1, len(normalizedSPY)):
    dailyReturnS = ((normalizedSPY[s] / normalizedSPY[s-1]) -1) * 100
    dailySPYReturnList.append(dailyReturnS)
    
# Average
avgTLT = np.average(dailyTLTReturnList)
avgSPY = np.average(dailySPYReturnList)
print("Average for TLT:", avgTLT)
print("Average for SPY:", avgSPY)

# Minimum
minTLT = np.min(dailyTLTReturnList)
minSPY = np.min(dailySPYReturnList)
print("Minimum for TLT:", minTLT)
print("Minimum for SPY:", minSPY)

# Maximum
maxTLT = np.max(dailyTLTReturnList)
maxSPY = np.max(dailySPYReturnList)
print("Maximum for TLT:", maxTLT)
print("Maximum for SPY:", maxSPY)
```
