```python
# Importing libraries to use
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import quandl
```

### Question 1

```python
malnutritionPrev = pd.read_excel('maln.xls', sheet_name = 'Data', skiprows = 3)    # Reading malnutrition data from excel file
gdpPerCapita = pd.read_excel('gdp.xls', sheet_name = 'Data', skiprows = 3)    # Reading malnutrition data from excel file

allTitles = malnutritionPrev.columns                # List of all titles name: use malnutrition or gdp since titles are similar 

# Plotting: Defining figure properties
plt.figure(figsize = (12, 5), edgecolor = 'r', facecolor = 'y', linewidth = 1)

# Adding the title and labels to the plot
plt.title("MALNUTRITION PREVALENCE AGAINST GDP PER CAPITA (1960-2021, ALL COUNTRIES)")
plt.xlabel("GDP PER CAPITA")
plt.ylabel("MALNUTRITION PREVALENCE") 

# Looping through all rows of both datasets plotting only columns from 1960 to 2021
for i in range(4, (len(allTitles) - 1)):                                        # indices 4-length of allTitles list
    
# plotting corresponding columns for both datasets
    plt.scatter(gdpPerCapita[allTitles[i]], malnutritionPrev[allTitles[i]], color = 'blue')

# Showing the plot
plt.show()
```

```python
# Reading countries metadata for both malnutrition and GDP from excel file's sheets
malnMeta = pd.read_excel('maln.xls', sheet_name = 'Metadata - Countries')
gdpMeta = pd.read_excel('gdp.xls', sheet_name = 'Metadata - Countries')  

# Merging the original data sets with their correspondig metadata on country code
malnMetaMerge = pd.merge(malnutritionPrev, malnMeta, on = 'Country Code')
gdpMetaMerge = pd.merge(gdpPerCapita, gdpMeta, on = 'Country Code')

allTitls = gdpMetaMerge.columns             # List of all titles name: use malnutrition or gdp since titles are similar

# Plotting: Defining figure properties
plt.figure(figsize = (14, 7), edgecolor = 'r', facecolor = 'y', linewidth = 1)

# Adding the title and labels to the plot
plt.title("MALNUTRITION PREVALENCE VS GDP PER CAPITA (1960-2021, SIX REGIONS)")
plt.xlabel("GDP PER CAPITA")
plt.ylabel("MALNUTRITION PREVALENCE")


################################################################################################################################
# DATA FOR EAST ASIA & PACIFIC: Selecting only the rows with specified region
eapMALN =  malnMetaMerge[malnMetaMerge['Region'] == 'East Asia & Pacific'] 
eapGDP = gdpMetaMerge[gdpMetaMerge['Region'] == 'East Asia & Pacific']

for eap in range(4, (len(allTitls) - 4)): # Loop through the titles to take only year numbers titles
    
    # plotting corresponding columns(year numbers) for both GDP and Malnutrition of East Asia & Pacific
    eapPlot = plt.scatter(eapGDP[allTitls[eap]], eapMALN[allTitls[eap]], color = 'blue', label = 'East Asia & Pacific')
    
    
################################################################################################################################
# DATA FOR EUROPE & CENTRAL ASIA: Selecting only the rows with specified region
ecaMALN =  malnMetaMerge[malnMetaMerge['Region'] == 'Europe & Central Asia']
ecaGDP = gdpMetaMerge[gdpMetaMerge['Region'] == 'Europe & Central Asia']

for eca in range(4, (len(allTitls) - 4)): # Loop through the titles to take only year numbers titles
    
    # plotting corresponding columns(year numbers) for both GDP and Malnutrition of Europe & Central Asia
    ecaPlot = plt.scatter(ecaGDP[allTitls[eca]], ecaMALN[allTitls[eca]], color = 'red', label = 'Europe & Central Asia')
    

################################################################################################################################
# DATA FOR LATIN AMERICA & CARIBBEAN: Selecting only the rows with specified region
lacMALN =  malnMetaMerge[malnMetaMerge['Region'] == 'Latin America & Caribbean']
lacGDP = gdpMetaMerge[gdpMetaMerge['Region'] == 'Latin America & Caribbean']
 
for lac in range(4, (len(allTitls) - 4)): # Loop through the titles to take only year numbers titles
    
    # plotting corresponding columns(year numbers) for both GDP and Malnutrition of Latin America & Caribbean
    lacPlot = plt.scatter(lacGDP[allTitls[lac]], lacMALN[allTitls[lac]], color = 'green', label = 'Latin America & Caribbean')

    
################################################################################################################################
# DATA FOR MIDDLE EAST & NORTH AFRICA: Selecting only the rows with specified region
menaMALN =  malnMetaMerge[malnMetaMerge['Region'] == 'Middle East & North Africa']
menaGDP = gdpMetaMerge[gdpMetaMerge['Region'] == 'Middle East & North Africa']

for mena in range(4, (len(allTitls) - 4)): # Loop through the titles to take only year numbers titles
    
    # plotting corresponding columns(year numbers) for both GDP and Malnutrition of Middle East & North Africa
    menaPlot = plt.scatter(menaGDP[allTitls[mena]], menaMALN[allTitls[mena]], color = 'gray', label = 'Middle East & North Africa')
    
    
################################################################################################################################
# DATA FOR SOUTH ASIA: Selecting only the rows with specified region
saMALN =  malnMetaMerge[malnMetaMerge['Region'] == 'South Asia']
saGDP = gdpMetaMerge[gdpMetaMerge['Region'] == 'South Asia']

for sa in range(4, (len(allTitls) - 4)): # Loop through the titles to take only year numbers titles
    
    # plotting corresponding columns(year numbers) for both GDP and Malnutrition of South Asia
    saPlot = plt.scatter(saGDP[allTitls[sa]], saMALN[allTitls[sa]], color = 'yellow', label = 'South Asia')
    
    
################################################################################################################################
# DATA FOR SUB-SAHARAN AFRICA: Selecting only the rows with specified region
ssaMALN =  malnMetaMerge[malnMetaMerge['Region'] == 'Sub-Saharan Africa']
ssaGDP = gdpMetaMerge[gdpMetaMerge['Region'] == 'Sub-Saharan Africa']

for ssa in range(4, (len(allTitls) - 4)): # Loop through the titles to take only year numbers titles
    
    # plotting corresponding columns(year numbers) for both GDP and Malnutrition of Sub-Saharan Africa
    ssaPlot = plt.scatter(ssaGDP[allTitls[ssa]], ssaMALN[allTitls[ssa]], color = 'black', label = 'Sub-Saharan Africa')
    
    
################################################################################################################################

# Adding the legend to each of the region scatter plot data
plt.legend([eapPlot, ecaPlot, lacPlot, menaPlot, saPlot, ssaPlot],
           ['East Asia & Pacific','Europe & Central Asia', 'Latin America & Caribbean', 'Middle East & North Africa', 'South Asia', 'Sub-Saharan Africa'])

# Showing the plot
plt.show()
```

```python
# Plotting: Defining figure properties
plt.figure(figsize = (14, 7), edgecolor = 'r', facecolor = 'y', linewidth = 1)

# Adding the title and labels to the plot
plt.title('MALNUTRITION PREVALENCE VS GDP PER CAPITA (1960-2021, FOUR INCOME LEVELS)')
plt.xlabel('GDP PER CAPITA')
plt.ylabel('MALNUTRITION PREVALENCE')

################################################################################################################################
# DATA FOR HIGH INCOME LEVEL: Selecting only the rows with specified income level
hiMALN =  malnMetaMerge[malnMetaMerge['IncomeGroup'] == 'High income']
hiGDP = gdpMetaMerge[gdpMetaMerge['IncomeGroup'] == 'High income']

for hi in range(4, (len(allTitls) - 4)): # Loop through the titles to take only year numbers titles
    
    # plotting corresponding columns for both GDP and Malnutrition for High income level
    hiPlot = plt.scatter(hiGDP[allTitls[hi]], hiMALN[allTitls[hi]], color = 'blue', label = 'High income')

################################################################################################################################
# DATA FOR LOW INCOME LEVEL: Selecting only the rows with specified income level
liMALN =  malnMetaMerge[malnMetaMerge['IncomeGroup'] == 'Low income']
liGDP = gdpMetaMerge[gdpMetaMerge['IncomeGroup'] == 'Low income']

for li in range(4, (len(allTitls) - 4)): # Loop through the titles to take only year numbers titles
    
    # plotting corresponding columns for both GDP and Malnutrition for Low income level
    liPlot = plt.scatter(liGDP[allTitls[li]], liMALN[allTitls[li]], color = 'brown', label = 'Low income')
    
################################################################################################################################
# DATA FOR LOWER MIDDLE INCOME LEVEL: Selecting only the rows with specified income level
lmiMALN =  malnMetaMerge[malnMetaMerge['IncomeGroup'] == 'Lower middle income']
lmiGDP = gdpMetaMerge[gdpMetaMerge['IncomeGroup'] == 'Lower middle income']

for lmi in range(4, (len(allTitls) - 4)): # Loop through the titles to take only year numbers titles
    
    # plotting corresponding columns for both GDP and Malnutrition for Lower middle income level
    lmiPlot = plt.scatter(lmiGDP[allTitls[lmi]], lmiMALN[allTitls[lmi]], color = 'yellow', label = 'Lower middle income')
    
################################################################################################################################
# DATA FOR UPPER MIDDLE INCOME LEVEL: Selecting only the rows with specified income level
umiMALN =  malnMetaMerge[malnMetaMerge['IncomeGroup'] == 'Upper middle income']
umiGDP = gdpMetaMerge[gdpMetaMerge['IncomeGroup'] == 'Upper middle income']

for umi in range(4, (len(allTitls) - 4)): # Loop through the titles to take only year numbers titles
    
    # plotting corresponding columns for both GDP and Malnutrition for Lower middle income level
    umiPlot = plt.scatter(umiGDP[allTitls[umi]], umiMALN[allTitls[umi]], color = 'aqua', label = 'Upper middle income')
    
################################################################################################################################

# Adding the legend to each of the income levels scatter plot data
plt.legend([hiPlot, liPlot, lmiPlot, umiPlot], ['High income', 'Low income', 'Lower middle income', 'Upper middle income'])

# Showing the plot
plt.show()
```

### Question 3

```python
co2Emissions = pd.read_excel('CO2.xls', sheet_name = 'Data', skiprows = 3)    # Reading the CO2 data from excel file
pEnrollment = pd.read_excel('primary.xls', sheet_name = 'Data', skiprows = 3) # Reading the enrollment data from excel file

# Extracting only Country Name and the year 2010 to use
co2EmExtracted = co2Emissions[['Country Name', '2010']]
pEnrExtracted = pEnrollment[['Country Name', '2010']]

# CO2 Emissions Summary Statistics
co2Mean = co2EmExtracted['2010'].mean(skipna = True)
co2Median = co2EmExtracted['2010'].median(skipna = True)
co2Std = co2EmExtracted['2010'].std(skipna = True)
co2percentile5 = co2EmExtracted['2010'].quantile(0.05)
co2percentile25 = co2EmExtracted['2010'].quantile(0.25)
co2percentile75 = co2EmExtracted['2010'].quantile(0.75)
co2percentile95 = co2EmExtracted['2010'].quantile(0.95)

# Primary Enrollment Summary Statistics
pEnrolMean = pEnrExtracted['2010'].mean(skipna = True)
pEnrolMedian = pEnrExtracted['2010'].median(skipna = True)
pEnrolStd = pEnrExtracted['2010'].std(skipna = True)
pEnrolpercentile5 = pEnrExtracted['2010'].quantile(0.05)
pEnrolpercentile25 = pEnrExtracted['2010'].quantile(0.25)
pEnrolpercentile75 = pEnrExtracted['2010'].quantile(0.75)
pEnrolpercentile95 = pEnrExtracted['2010'].quantile(0.95)

# Creating summary statistics table for both CO2 emissions and Primary schools enrolment
co2Summary = pd.DataFrame({'CO2 Emissions  (metric tons/capita) 2010 Summary': [co2Mean, co2Median, co2Std, co2percentile5, co2percentile25, co2percentile75, co2percentile95]}, 
                 index = ['Mean', 'Median', 'Standard Deviation', '5 Percentile', '25 Percentile', '75 Percentile', '95 Percentile'])
pEnrolSummary = pd.DataFrame({'School enrolment, primary (% net) 2010 Summary': [pEnrolMean, pEnrolMedian, pEnrolStd, pEnrolpercentile5, pEnrolpercentile25, pEnrolpercentile75, pEnrolpercentile95]}, 
                 index = ['Mean', 'Median', 'Standard Deviation', '5 Percentile', '25 Percentile', '75 Percentile', '95 Percentile'])

# Displaying the tables
display(co2Summary)
display(pEnrolSummary)
```

### Question 4

```python
fertilityRate = pd.read_excel('fertility.xls', sheet_name = 'Data', skiprows = 3)  # Reading fertility data from excel file
gdpPerCapita = pd.read_excel('gdp.xls', sheet_name = 'Data', skiprows = 3)         # Reading gdp data from excel file

# Extracting only Country Name and the year 2010 to use
fertExtracted = fertilityRate[['Country Name', '2010']]
gdpExtracted= gdpPerCapita[['Country Name', '2010']]

# Plotting: Defining figure properties
plt.figure(figsize = (14, 7), edgecolor = 'r', facecolor = 'y', linewidth = 1)

# Adding the title and labels to the plot
plt.title("FERTILITY RATE AGAINST GDP PER CAPITA FOR ALL COUNTRIES IN 2010")
plt.xlabel("GDP PER CAPITA")
plt.ylabel("FERTILITY RATE")

# Plotting Fertility rate versus GDP per capita for all countries in 2010
plt.scatter(gdpExtracted['2010'], fertExtracted['2010'], marker = 'o', color = 'blue')

# Showing the plot
plt.show()
```

```python
# Extracting only Country Name and the year 2010 and 1990 to use
fertExtracted1990 = fertilityRate[['Country Name', '1990']]
fertExtracted2010 = fertilityRate[['Country Name', '2010']]

# Sorting the data to use while plotting the distribution
fertSorted1990 = np.sort(fertExtracted1990['1990'])
fertSorted2010 = np.sort(fertExtracted2010['2010'])

# Finding the fertility rate's cumulative distribution functions for 1990 & 2010
fertCDF1990 = np.arange(len(fertExtracted1990)) / (len(fertExtracted1990) - 1)
fertCDF2010 = np.arange(len(fertExtracted2010)) / (len(fertExtracted2010) - 1)

# Finding the mean
mean1990 = fertExtracted1990['1990'].mean()
mean2010 = fertExtracted2010['2010'].mean()

# Finding the median
median1990 = fertExtracted1990['1990'].median()
median2010 = fertExtracted2010['2010'].median()

# Plotting: Defining figure properties
plt.figure(figsize = (14, 8), edgecolor = 'r', facecolor = 'y', linewidth = 1)

# Plotting on the same axes cumulative distribution functions for the fertility rate variable using 1990 & 2010 data
plt.plot(fertSorted1990, fertCDF1990, color = 'black', label = 'Year 1990 distribution')
plt.plot(fertSorted2010, fertCDF2010, color = 'green', label = 'Year 2010 distribution')

# Plotting the mean using vertical lines 
plt.axvline(x = mean1990, color = 'red', label = 'Mean for 1990', linestyle = '-', linewidth = 1.5, ymin = 0.02, ymax = 0.97)
plt.axvline(x = mean2010, color = 'blue', label = 'Mean for 2010', linestyle = '-', linewidth = 1.5, ymin = 0.02, ymax = 0.97)

# Plotting the median using vertical lines 
plt.axvline(x = median1990, color = 'pink', label = 'Median for 1990', linestyle = '-', linewidth = 1.5, ymin = 0.02, ymax = 0.97)
plt.axvline(x = median2010, color = 'aqua', label = 'Median for 2010', linestyle = '-', linewidth=1.5, ymin = 0.02, ymax = 0.97)

# Adding the title and labels to the plot
plt.title('CUMULATIVE DISTRIBUTION FUNCTIONS FOR THE FERTILITY RATE IN 1990 & 2010')
plt.ylabel('CUMULATIVE DISTRIBUTION')
plt.xlabel('FERTILITY RATE')

# Adding legend labels to the plot
plt.legend()

# Showing the plot
plt.show()
```

### Question 5

```python
HPI2016Data = pd.read_excel('hpi-data-2016.xlsx', sheet_name = 'Complete HPI data', skiprows = 5) # Reading HPI data from excel file
CPI2016Data = pd.read_excel('CPI2016_Results.xlsx', sheet_name = 'CPI2016_FINAL_16Jan') # Reading CPI data from excel file

HPICPIData = pd.merge(HPI2016Data, CPI2016Data, on = 'Country') # Merging the HPI and CPI datasets on country names 

# Extracting the rankings for both CPI and HPI from the merged dataset
hpiRanking = HPICPIData['HPI Rank']
cpiRanking = HPICPIData['Rank']

# Plotting: Defining figure properties
plt.figure(figsize = (15, 9), edgecolor = 'r', facecolor = 'y', linewidth = 1)

# Adding the title and labels to the plot
plt.title("RELATIONSHIP BETWEEN HPI AND CPI BY COUNTRIES RANKS")
plt.ylabel("HAPPY PLANET INDEX")
plt.xlabel("CORRUPTION PERCEPTION INDEX")

# Plotting the dots using the ranks for every country code
plt.scatter(cpiRanking, hpiRanking, color = 'green', s = 100, marker = 'o')

# Returning the count(index) and value(countryCode) from HPICPIData['WB Code'] column for every of its elements
for index, countryCode in enumerate(HPICPIData['WB Code']):
    plt.annotate(countryCode, (cpiRanking[index], hpiRanking[index]))  # Using country's abbreviations to place annotations

# Showing the plot
plt.show()
```
