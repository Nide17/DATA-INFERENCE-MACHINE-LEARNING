### Question 1

```python
foldsNumber = 0                                           # Initializing the number times the paper is folded
foldedThickness = 0                                       # Initializing the folded paper thickness

# The height of Mount Everest is 8,848 m which is equal to 8,848,000 mm

while foldedThickness <= 8848000:                         # Looping until the folded paper's thickness <= Everest height
    foldsNumber = foldsNumber + 1                         # Incrementing the folds number while looping
    foldedThickness = pow(2, foldsNumber)                 # As the paper is folded its thickness becomes by 2 power fold number
    
print("It would take", foldsNumber, "folds to exceed the height of Mount Everest")   # Printing the final required folds number 
```

### Question 2

##### Steps to solve the question's equation

V(t) = V(1) exp(-at)

__WHERE__: a = -0.1

__AT HALF__: V(t) = V(1) / 2

__NOW__: V(1)/2 = V(1) exp(-at)

1/2 = exp(-0.1t)

exp(-0.1t) = 1/2

__BECAUSE: exp(x) = e^x__

__THUS__: e^-0.1t = 1/2

ln(e^-0.1t) = ln(1/2)

-0.1t ln(e) = ln(1/2)

-0.1t = ln(1/2)

__THEREFORE:__ t = (ln(1/2)) / (-0.1)

```python
# Applying the final equation and printing the answer
t = np.log(1 / 2) / (-0.1)
print("It would take", t, "time for the volume to decrease to less than one half of its initial volume")
```

### Question 3

```python
# Compound interest formula: A = P(1+(r/n))**nt
P = 100                                               # The principal amount
r = 5 / 100                                           # Interest rate of 5%
n = 1                                                 # Frequency is 1 since the annualized interest rate is compounded annually

# Finding the amount after every number of years: amount = P * pow((1+r/n), n * t) 
for t in [1, 2, 3, 4, 5]:
    amount= P * pow((1 + (r / n)), (n * t))                # Computing the amount on each iteration of years
    print("The amount at year", t, "is", round(amount))
```

### Question 4

```python
#  Loan payment formula: M = Pr / (1 −(1+r) ** −n)

# M: is the monthly payment
# P: is the original principal amount
# r: is the interest rate per month (annual interest divided by 12)
# n: is the number of months to repay the loan or total number of payments

P = 20000
r = 0.01 # the 1% interest rate
index = 0

for yearNbr in [1, 2, 3]:
    M = round((P * r) / (1 - (pow((1 + r), (-12 * yearNbr))))) # Applying the "M = Pr / (1 −(1+r) ** −n)" formula
    index = index + 1
    print("The payment in", index, "year(s) is", M)
```

### Question 5

```python
initInvest = 100000
custNbr = 100

pymts = 0
cPymtsList = []                                  # Initializing a list to store cumulated payments each day

daysNbr = 0
daysList = []                                    # List to store day numbers

while pymts <= initInvest:                       # iterate until the payments is greater than or equal to amount invested
    custNbr = custNbr + (custNbr * (1 / 100))    # updating customers
    daysNbr = daysNbr + 1                        # updating days
    daysList.append(daysNbr)                     # Keeping track of the number of days
    pymts = pymts + custNbr * 10                 # updating the payments
    cPymtsList.append(pymts)                     # Keeping track of cumulated profit for each day
    
print("It would take",  daysNbr, "to repay the initial investment")
```

```python
# Plotting cumulated profits per day

plt.figure(figsize = (11, 5), edgecolor = 'r', facecolor = 'y', linewidth = 1)
plt.plot(daysList, cPymtsList, color = 'green')

#  Adding the title and labels to the plot
plt.title("CUMULATED PROFITS PER DAY")
plt.xlabel("Days")
plt.ylabel("Payments") 

# Finding the break even point day = where the initial investment is repaid fully
profitDay = np.interp(initInvest, cPymtsList , daysList)               

plt.scatter(profitDay, initInvest, color = 'red', s = 100, marker = '*')  # Showing the break even point on the plot
plt.annotate("Break even day", xy = (68.5, 100000),  xytext = (50, 102000),  arrowprops = dict(arrowstyle = "fancy", connectionstyle = "arc3"))

plt.show()
```

### Question 6

```python
ebolaExcelData = pd.read_excel('ebola_download.xls')                       # Reading the data from excel file
ebolaExtractedData = ebolaExcelData[['Date', 'Cases', 'Death']]            # Extracting Dates, Cases, and Deaths from excel data
ebolaExtractedData.info()                                                  # Getting info about the data 
```

```python
# Get missing dates by generating dates from the first date to the last date in the dataset
ebolaSortedData = ebolaExtractedData.sort_values(by = 'Date')               # Sorting the data by date first
firstDate = ebolaSortedData['Date'][0]                                      # The first date in the sorted data
lastDate = ebolaSortedData['Date'][len(ebolaExtractedData['Date']) - 1]     # The last date in the sorted data

allDates = pd.date_range(start = firstDate, end = lastDate)                 # Creating a date range including also skipped dates
len(allDates)                                                               # View the length or amount of total dates
```

```python
ebolaSortedData = ebolaSortedData.set_index('Date')                          # Set the 'Date' column to be the index of dataset

# Integrate the missing dates into dataset. Conform dataset to new index. Placing NaN where no value exist 
ebolaReindexedData = ebolaSortedData.reindex(allDates)  
ebolaReindexedData.isnull().sum()                                            # Counting empty fields
```

```python
# Interpolating linearly to get NaN (missing) values from the dataset for cases and deaths
ebolaInterpolatedData = ebolaReindexedData.interpolate()  

ebolaInterpolatedData = ebolaInterpolatedData.rename_axis('Date').reset_index() # Adding the title to the index (date) column
ebolaInterpolatedData.isnull().sum()                                            # Counting  empty fields
```

#### Extracting indices (first index) where case values are greater than 100, 500, 1000, 2000, 5000

##### Ebola cases values

```python
casesList = [100, 500, 1000, 2000, 5000]                # The given threshold numbers
i = 0
excCases = []                                           # The list to store the first case numbers exceeding the thresholds

while i < len(casesList):                               # Loop through the threshold numbers
    for c in ebolaInterpolatedData['Cases']:            # Finding the first number exceding the current threshold number
        if c >= casesList[i]: 
            break                                       # If the number is found, the loop stops and go to outer loop again
    excCases.append(c)                                  # Storing the first exceeding case numbers to the list
    i = i + 1
    
excCases                                                # Displaying the first case numbers to exceed the threshold numbers
```

##### Ebola cases dates values

```python
c = 0
excCasesDates = []                                      # The list to store the dates the case exceed the thresholds

# Finding the dates the numbers of cases exceeded the threshold
while c < len(excCases):                                              
    caseDate = ebolaInterpolatedData[ebolaInterpolatedData['Cases'] == excCases[c]]  # If the exceeding number exist, store its
    caseDate = caseDate['Date']                          # Extracting the date column
    
    print("The date", c+1, "for Ebola cases is: ", caseDate.iloc[0])  # Printing 5 dates of ebola cases
    excCasesDates.append(caseDate)                       # Updating excCasesDates list
    c = c + 1
```

##### Ebola deaths values

```python
deathsList = [100, 500, 1000, 2000, 5000]                 # The given threshold numbers
i = 0
excDeaths = []                                            # The list to store the first death numbers exceeding the thresholds

while i < len(deathsList):                                # Loop through the threshold numbers
    for c in ebolaInterpolatedData['Death']:              # Finding the first number exceding the current threshold number
        if c >= deathsList[i]: 
            break                                         # If the number is found, the loop stops and go to outer loop again
    excDeaths.append(c)                                   # Storing the first exceeding death numbers to the list
    i = i + 1
    
excDeaths                                                 # Displaying the first death numbers to exceed the threshold numbers
```

##### Ebola cases dates values

```python
d = 0
excDeathsDates = []                                        # The list to store the dates the case exceed the thresholds

# Finding the dates the numbers of deaths exceeded the threshold
while d < len(excDeaths):
    deathDate = ebolaInterpolatedData[ebolaInterpolatedData['Death'] == excDeaths[d]]  # If the exceeding number exist, store its
    deathDate = deathDate['Date']                          # Extracting the date column
    
    print("The date", d + 1, "for Ebola death is: ", deathDate.iloc[0])                  # Printing 5 dates of ebola deaths
    excDeathsDates.append(deathDate)                       # Updating excDeathsDates list
    d = d + 1
```

#### Plotting cases vs dates

```python
# Extracting the cases and deaths to be plotted
ebolaCasesData = ebolaInterpolatedData["Cases"]
ebolaDeathsData = ebolaInterpolatedData["Death"]

# Plotting cases and deaths against dates
plt.figure(figsize = (11, 5), edgecolor = 'r', facecolor = 'y', linewidth = 1)

plt.plot(ebolaInterpolatedData["Date"], ebolaCasesData, label = 'Cases', color = 'green')
plt.scatter(excCasesDates, excCases, marker = 'o', color = 'red')

plt.plot(ebolaInterpolatedData["Date"], ebolaDeathsData, label = 'Deaths', color = 'orange')
plt.scatter(excDeathsDates, excDeaths, marker = 'o', color = 'red')

plt.title("EBOLA CASES AND DEATHS AGAINST DATES")
plt.xlabel("Dates")
plt.ylabel("Values of Ebola Cases and Deaths")
plt.legend()
plt.show()
```

### Question 7

```python
interpolatedCases = ebolaInterpolatedData['Cases']              # Getting interpolated cases from question 6
interpolatedDeaths = ebolaInterpolatedData['Death']             # Getting interpolated deaths from question 6
```

```python
# Computing growth rates: Formula: GR = (Absolute change / Original value) * 100 => Absolute change = New value - Original value

casesGR = 0
c = 1 

while c <= (len(interpolatedCases) - 1):
    dailyGR = ((interpolatedCases[c] - interpolatedCases[c-1]) / interpolatedCases[c-1]) * 100     # The daily growth rate
    casesGR = casesGR + dailyGR
    c = c + 1

casesAvgGR = casesGR / (len(interpolatedCases) - 1)             # The average growth rate for cases

print("The average growth rate per day for cases is:", casesAvgGR)
```

```python
# Computing growth rates: Formula: GR = (Absolute change / Original value) * 100 => Absolute change = New value - Original value
deathsGR = 0
d = 1 

while d <= (len(interpolatedDeaths) - 1):
    dailyGR = ((interpolatedDeaths[d] - interpolatedDeaths[d-1]) / interpolatedDeaths[d-1]) * 100
    deathsGR = deathsGR + dailyGR
    d = d + 1

deathsAvgGR = deathsGR / (len(interpolatedDeaths) - 1)           # The average growth rate for deaths

print("The average growth rate per day for death is:" ,deathsAvgGR)
```

### Question 9

```python
# Reading data of stocks
tlt = pd.read_csv('TLT.csv')
spy = pd.read_csv('SPY.csv')

# Extracting dates and adjusted closing values from data
extractedTLT = tlt[['Date', 'Adj Close']]
extractedSPY = spy[['Date', 'Adj Close']] 
```

```python
#  Extracting the 'Adj Close' column from both stocks
extractedAdjCloseTLT = extractedTLT['Adj Close']
extractedAdjCloseSPY = extractedSPY['Adj Close']

# Lists to store the normalized data
normalizedTLT = []
normalizedSPY = []

# Normalizing so that the first element in the TLT series is 100
for tlt in extractedAdjCloseTLT:
    normalized = (tlt / extractedAdjCloseTLT[0]) * 100
    normalizedTLT.append(normalized)
    
# Normalizing so that the first element in the SPY series is 100
for spy in extractedAdjCloseSPY:
    normalized = (spy / extractedAdjCloseSPY[0]) * 100
    normalizedSPY.append(normalized)
```

```python
# Plotting
plt.figure(figsize = (11, 5), edgecolor = 'r', facecolor = 'y', linewidth = 1)
extractedTLTDate = pd.to_datetime(extractedTLT['Date'])                  # converting the timestamps to datetime
extractedSPYDate = pd.to_datetime(extractedSPY['Date'])

plt.xlim(pd.Timestamp('2014-01-01'), pd.Timestamp('2015-08-31'))         # set time limits of the Date axis.

plt.plot(extractedTLTDate, normalizedTLT, label = 'TLT', color = 'green')
plt.plot(extractedSPYDate, normalizedSPY, label = 'SPY', color = 'orange')

plt.title("COMPARISON BETWEEN TLT AND SPY")
plt.ylabel("TLT and SPY")
plt.xlabel("Date")
plt.legend()
plt.show()
```

### Question 10

```python
# Returns formula: r(t)= (p(t)/p(t-1))-1   => 

dailyTLTReturnList = []
dailySPYReturnList = []

# Finding and storing daily returns for TLT
for t in range(1, len(extractedAdjCloseTLT)):
    dailyReturnT = ((extractedAdjCloseTLT[t] / extractedAdjCloseTLT[t-1]) -1) * 100
    dailyTLTReturnList.append(dailyReturnT)

# Finding and storing daily returns for SPY
for s in range(1, len(extractedAdjCloseSPY)):
    dailyReturnS = ((extractedAdjCloseSPY[s] / extractedAdjCloseSPY[s-1]) -1) * 100
    dailySPYReturnList.append(dailyReturnS)
    
# Average
avgTLT = np.average(dailyTLTReturnList)
avgSPY = np.average(dailySPYReturnList)
print("Average for TLT:", avgTLT)
print("Average for SPY:", avgSPY)

# Minimum
minTLT = np.min(dailyTLTReturnList)
minSPY = np.min(dailySPYReturnList)
print("Minimum for TLT:", minTLT)
print("Minimum for SPY:", minSPY)

# Maximum
maxTLT = np.max(dailyTLTReturnList)
maxSPY = np.max(dailySPYReturnList)
print("Maximum for TLT:", maxTLT)
print("Maximum for SPY:", maxSPY)
```
